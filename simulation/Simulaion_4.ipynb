```python
# Based on the sample_size = 400 power tables at size-adjusted levels ≈ 0.05 and ≈ 0.10 shown in your image [file:32].

import matplotlib.pyplot as plt
import numpy as np

# =========================
# Data: sample_size = 400
# =========================
alpha_x = np.array([0.05, 0.08, 0.10, 0.12, 0.15, 0.20])

# Size-adjusted level ≈ 0.05
power_005 = {
    "Baseline":            [0.14, 0.17, 0.36, 0.55, 0.79, 0.99],
    "Mean Alignment 0.2":  [0.08, 0.15, 0.38, 0.60, 0.84, 1.00],
    "Mean Alignment 0.3":  [0.11, 0.34, 0.48, 0.65, 0.84, 0.99],
    "Mean Alignment 0.5":  [0.09, 0.31, 0.38, 0.56, 0.87, 1.00],
}

# Size-adjusted level ≈ 0.10
power_010 = {
    "Baseline":            [0.20, 0.31, 0.41, 0.67, 0.80, 0.99],
    "Mean Alignment 0.2":  [0.21, 0.36, 0.49, 0.73, 0.92, 1.00],
    "Mean Alignment 0.3":  [0.26, 0.58, 0.62, 0.78, 0.91, 1.00],
    "Mean Alignment 0.5":  [0.13, 0.38, 0.46, 0.72, 0.89, 1.00],
}

# =========================
# Style settings
# =========================
style_map = {
    "Baseline": {
        "color": "#2f2f2f",
        "marker": "o",
        "linestyle": "-",
        "linewidth": 2.2,
        "markersize": 6
    },
    "Mean Alignment 0.2": {
        "color": "#d55e00",
        "marker": "s",
        "linestyle": "--",
        "linewidth": 2.2,
        "markersize": 6
    },
    "Mean Alignment 0.3": {
        "color": "#0072b2",
        "marker": "^",
        "linestyle": "-.",
        "linewidth": 2.2,
        "markersize": 6
    },
    "Mean Alignment 0.5": {
        "color": "#009e73",
        "marker": "D",
        "linestyle": ":",
        "linewidth": 2.6,
        "markersize": 6
    },
}

plt.style.use("seaborn-v0_8-whitegrid")

# Bigger figure so overlapping parts are clearer
fig, axes = plt.subplots(1, 2, figsize=(16, 6.5), dpi=180, sharey=True)

# =========================
# Left panel: level ≈ 0.05
# =========================
ax = axes[0]
for name, y in power_005.items():
    st = style_map[name]
    ax.plot(
        alpha_x, y,
        label=name,
        color=st["color"],
        marker=st["marker"],
        linestyle=st["linestyle"],
        linewidth=st["linewidth"],
        markersize=st["markersize"],
        markerfacecolor=st["color"],
        markeredgecolor="white",
        markeredgewidth=0.8,
        alpha=0.95
    )

ax.set_title(r"Statistical Power ($sample\_size = 400$, Size-Adj. Level $\approx 0.05$)", fontsize=13)
ax.set_xlabel(r"Dependence Strength $\alpha_x$", fontsize=12)
ax.set_ylabel("Size-Adjusted Power", fontsize=12)
ax.set_xlim(0.045, 0.205)
ax.set_ylim(-0.02, 1.04)
ax.set_xticks(alpha_x)
ax.set_yticks(np.arange(0.0, 1.01, 0.2))
ax.tick_params(axis="both", labelsize=10)
ax.grid(True, linestyle="--", linewidth=0.6, alpha=0.45)

legend = ax.legend(
    loc="upper left",
    fontsize=8,
    frameon=True,
    ncol=1,
    borderpad=0.5,
    labelspacing=0.35,
    handlelength=2.5,
    handletextpad=0.6
)
legend.get_frame().set_alpha(0.92)
legend.get_frame().set_edgecolor("#bbbbbb")

# =========================
# Right panel: level ≈ 0.10
# =========================
ax = axes[1]
for name, y in power_010.items():
    st = style_map[name]
    ax.plot(
        alpha_x, y,
        label=name,
        color=st["color"],
        marker=st["marker"],
        linestyle=st["linestyle"],
        linewidth=st["linewidth"],
        markersize=st["markersize"],
        markerfacecolor=st["color"],
        markeredgecolor="white",
        markeredgewidth=0.8,
        alpha=0.95
    )

ax.set_title(r"Statistical Power ($sample\_size = 400$, Size-Adj. Level $\approx 0.10$)", fontsize=13)
ax.set_xlabel(r"Dependence Strength $\alpha_x$", fontsize=12)
ax.set_xlim(0.045, 0.205)
ax.set_ylim(-0.02, 1.04)
ax.set_xticks(alpha_x)
ax.tick_params(axis="both", labelsize=10)
ax.grid(True, linestyle="--", linewidth=0.6, alpha=0.45)

# =========================
# Final layout
# =========================
for ax in axes:
    for spine in ax.spines.values():
        spine.set_linewidth(0.8)

plt.subplots_adjust(wspace=0.08, left=0.07, right=0.98, top=0.88, bottom=0.13)

# Optional: save high-resolution figure
# plt.savefig("power_sample400_comparison.png", dpi=300, bbox_inches="tight")

plt.show()
```
